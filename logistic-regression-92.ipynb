```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv("/kaggle/input/brasilian-houses-to-rent/houses_to_rent.csv")
```

```python
data.head()
```

```python
data.drop("Unnamed: 0",axis = 1, inplace = True)
```

```python
data.head(2)
```

```python
data["floor"].replace(to_replace = "-",value = 0,inplace = True)
```

```python
data.head(2)
```

```python
data.info()
```

```python
data["animal"].replace(to_replace=("acept"),value = 1,inplace = True)
data["animal"].replace(to_replace=("not acept"),value = 0,inplace = True)
```

```python
data["furniture"].replace(to_replace=("furnished"),value = 1,inplace = True)
data["furniture"].replace(to_replace=("not furnished"),value = 0,inplace = True)
```

```python
data
```

```python
for _ in ["hoa","rent amount","property tax","fire insurance","total"]:
    data[_].replace(to_replace =("R\$"),value = "", regex = True , inplace = True)
    data[_].replace(to_replace =(","),value = "", regex = True , inplace = True)
    
```

```python
data
```

```python
data.describe()
```

```python
data
```

```python
data["hoa"].replace(to_replace=("Sem info"),value = 0, inplace = True)
data["hoa"].replace(to_replace=("Incluso"),value = 0, inplace = True)
```

```python
data["property tax"].replace(to_replace=("Incluso"),value = 0, inplace = True)
```

```python
data
```

```python
data = data.astype(dtype=np.int64)
```

```python
data.info()
```

```python
data = data.sample(frac = 1).reset_index(drop = True)
```

```python
data.head(2)
```

```python
y = data["city"]
x = data.drop("city", axis = 1)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```python
pd.DataFrame(x).head(10)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.7)
```

```python
x.size
```

```python
x.shape
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(penalty = "l2" , verbose = 1)
```

```python
model.fit(x_train, y_train)
```

```python
print(model.score(x_test, y_test))
```

```python
from sklearn.metrics import f1_score
```

```python
model_pred = model.predict(x_test)
```

```python
model_pred
```

```python
print(f1_score(model_pred, y_test))
```
